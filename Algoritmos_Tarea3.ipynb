## Ejercicio 1

Algoritmo Suma_Redondeada_Tres_Cifras

    Definir s1, s2, termino, resultado como reales
    Definir i como entero

    // Función para redondear a tres cifras significativas
    Función ROUND3(x)
        // Redondea el número x a tres cifras significativas
        devolver redondear_a_tres_cifras(x)
    FinFunción

    // Inicialización
    s1 ← 0
    s2 ← 0

    // --- Suma ascendente ---
    Para i ← 1 Hasta 10 Con paso 1 Hacer
        termino ← ROUND3(1 / (i^2))
        s1 ← ROUND3(s1 + termino)
    FinPara

    // --- Suma descendente ---
    Para i ← 10 Hasta 1 Con paso -1 Hacer
        termino ← ROUND3(1 / (i^2))

```python
# Ejercicio 1: sumas con aritmética de 3 dígitos
def round3(x):
    """Redondea a tres cifras significativas."""
    return float(f"{x:.3g}")

# (a) Suma de 1/i^2
# (b) Suma de 1/i^3

s1, s2 = 0.0, 0.0
# ascendente
for i in range(1, 11):
    s1 = round3(s1 + round3(1 / i**2))
# descendente
for i in range(10, 0, -1):
    s2 = round3(s2 + round3(1 / i**2))

print("Ascendente =", s1)
print("Descendente =", s2)
```

```python
# Ejercicio 2: Determinar n mínimo para error < tolerancia
from math import pi

def n_para_tolerancia(tol):
    n = 1
    while 4 / (2 * n + 1) > tol:
        n += 1
    return n

print("Para error < 1e-3: n =", n_para_tolerancia(1e-3))
print("Para error < 1e-10: n =", n_para_tolerancia(1e-5))
```

## Ejercicio 3

Algoritmo Error_Machin

    Definir pi, pi_aprox, error como real
    Definir n, k como entero

    pi ← 3.141592653589793   // Valor real de π

    // ---- Función para calcular arctan(x) mediante serie de Taylor ----
    Función ARCTAN_SERIES(x, n)
        suma ← 0
        Para k ← 1 Hasta n Hacer
            termino ← ((-1)^(k+1)) * (x^(2*k - 1)) / (2*k - 1)
            suma ← suma + termino
        FinPara
        devolver suma
    FinFunción

    // ---- Función principal que calcula el error de Machin ----
    Función MACHIN_ERROR(n)
        parte1 ← ARCTAN_SERIES(1/5, n)
        parte2 ← ARCTAN_SERIES(1/239, n)
        pi_aprox ← 4 * (4 * parte1 - parte2)
        error ← |pi - pi_aprox|
        devolver error
    FinFunción

    // ---- Programa principal ----
    Para n ← 1 Hasta 5 Hacer
        Mostrar "n =", n, " error =", MACHIN_ERROR(n)
    FinPara

FinAlgoritmo



```python
# Ejercicio 3: Determinar n mínimo para |error| < 1e-3 usando Machin
def machin_error(n):
    from math import pi
    def arctan_series(x, n):
        return sum(((-1)**(k+1)) * (x**(2*k-1)) / (2*k-1) for k in range(1, n+1))
    pi_aprox = 4 * (4 * arctan_series(1/5, n) - arctan_series(1/239, n))
    return abs(pi - pi_aprox)

for n in range(1, 6):
    print(f"n={n}, error={machin_error(n):.2e}")
```

## Ejercicio 1

INPUT: n, x[1..n]
S := 0
FOR i := n DOWNTO 1 DO
    S := S + x[i]
OUTPUT: S

```python
# Ejercicio 1: Sumar la serie finita Σx_i en orden inverso

def suma_inversa(x):
    """
    Suma los elementos del vector x en orden inverso.
    """
    S = 0
    for i in range(len(x) - 1, -1, -1):
        S += x[i]
    return S

# Ejemplo
x = [1, 2, 3, 4, 5]
print("Suma inversa =", suma_inversa(x))
```

## Ejercicio 2

INPUT: a,b,c
IF a = 0 THEN
    IF b = 0 THEN OUTPUT: "Sin solución o infinitas" ELSE OUTPUT: x = -c/b
ELSE
    D := b*b - 4*a*c
    r := sqrt(D)          # real o compleja
    q := -0.5*(b + sign(b)*r)
    x1 := q/a
    x2 := c/q             # estable numéricamente
    OUTPUT: x1, x2

```python
# Ejercicio 2: Cálculo estable de raíces cuadráticas

import cmath  # permite raíces complejas

def raices_cuadraticas(a, b, c):
    if a == 0:
        if b == 0:
            return "Sin solución o infinitas soluciones"
        else:
            return -c / b
    D = b**2 - 4*a*c
    r = cmath.sqrt(D)
    # Evitar cancelación
    q = -0.5 * (b + (b/abs(b) if b != 0 else 1) * r)
    x1 = q / a
    x2 = c / q
    return x1, x2

# Ejemplo
print("Raíces:", raices_cuadraticas(1, -3, 2))
```

## Ejercicio 3

INPUT: x, tol
S := 0
k := 1
REPEAT
    num := 2^(k-1)*x^(2^(k-1)-1) - 2^k*x^(2^k - 1)
    den := 1 - x^(2^(k-1)) + x^(2^k)
    S := S + num/den
    k := k + 1
UNTIL abs(S - (1+2*x)/(1+x+x^2)) < tol
OUTPUT: (k-1)   # número de términos

```python
# Ejercicio 3: Determinar cuántos términos se requieren para error < 1e-6

def serie(x, tol=1e-6):
    rhs = (1 + 2*x) / (1 + x + x**2)
    S = 0
    k = 1
    while True:
        num = 2**(k-1) * x**(2**(k-1) - 1) - 2**k * x**(2**k - 1)
        den = 1 - x**(2**(k-1)) + x**(2**k)
        S += num / den
        if abs(S - rhs) < tol:
            break
        k += 1
    return k, S, rhs

# Ejemplo
n, S, rhs = serie(0.25)
print(f"Se requieren {n} términos.")
print(f"Suma parcial = {S}, Valor real = {rhs}")
```
